```python
import matplotlib.pyplot as plt

# Sort data by funding and take top N entries
top_n = 10
top_df = df.sort_values(by="Fund allotted(in rs crores)", ascending=False).head(top_n)

# Bar Chart
plt.figure(figsize=(14, 10))
plt.barh(top_df["Department /Ministry"], top_df["Fund allotted(in rs crores)"], color='skyblue')
plt.xlabel('Fund allotted (in  crores)')
plt.ylabel('Department /Ministry')
plt.title(f'Top {top_n} Departments/Ministries by Fund Allotted')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest value on top
plt.tight_layout()
plt.show()
```

```python
# Bar Chart with Logarithmic Scale
plt.figure(figsize=(14, 10))
plt.barh(df["Department /Ministry"], df["Fund allotted(in rs crores)"], color='skyblue')
plt.xscale('log')
plt.xlabel('Fund allotted (in  crores) [Log Scale]')
plt.ylabel('Department /Ministry')
plt.title('Funds Allotted to Different Departments/Ministries')
plt.tight_layout()
plt.show()
```
